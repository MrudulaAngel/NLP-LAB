```python
#QUESTION 1
#read a paragraph and obtain the frequency of words
nltk.download('punkt')
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Sample paragraph
paragraph = "Natural Language Processing (NLP) is a subfield of artificial intelligence that focuses on the interaction between computers and humans using natural language."

# Tokenize the paragraph
words = word_tokenize(paragraph)

# Calculate word frequencies
freq_dist = FreqDist(words)

# Print word frequencies
print("Word frequencies:")
for word, frequency in freq_dist.items():
    print(f"{word}: {frequency}")
```

```python
#QUESTION 3
#read only the word content from web page and display their frequency
import requests
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# URL of the web page you want to analyze
url = "https://en.wikipedia.org/wiki/Natural_language_processing"

# Fetch the content from the web page
response = requests.get(url)
html_content = response.text

# Use BeautifulSoup to parse HTML and extract text
soup = BeautifulSoup(html_content, 'html.parser')
text_content = soup.get_text()

# Tokenize the text using NLTK
words = word_tokenize(text_content)

# Filter out non-alphabetic words (remove punctuation, numbers, etc.)
filtered_words = [word.lower() for word in words if word.isalpha()]

# Calculate word frequencies
freq_dist = FreqDist(filtered_words)

# Display word frequencies
print("Word frequencies:")
for word, frequency in freq_dist.items():
    print(f"{word}: {frequency}")
```

```python
#QUESTION 4
#plot the frequency for count obtained in eq3
# Plot the word frequencies
import matplotlib.pyplot as plt
plt.title('Word Frequency Distribution')
freq_dist.plot(40, cumulative=False)
plt.figure(figsize=(20, 10))
plt.show()
```
